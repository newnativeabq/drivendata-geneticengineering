```python
# Load
import pandas as pd
import os


data_dir = os.path.join(os.getcwd(), 'data', 'files')

os.listdir(data_dir)
```

```python
def _make_path(subd, fname):
    return os.path.join(subd, fname)

tval = pd.read_csv(_make_path(data_dir, 'train_values.csv'), index_col='sequence_id')
tlab = pd.read_csv(_make_path(data_dir, 'train_labels.csv'), index_col='sequence_id')

len(tval), len(tlab)
```

```python
# Since the data is 1:1, join everything to make plotting simpler
import numpy as np

subsample = np.random.choice(tlab.columns, 5)

tjf = tval.join(tlab[subsample])
tjf.head()
```

```python
# Many graphs are already available in benchmark here:
#     https://www.drivendata.co/blog/genetic-attribution-benchmark/

# Looking at distribution of three labs to see if fingerprinting or generation might be useful

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 5, figsize=(9,10), sharey=True)

subsets = []

for i, lab in enumerate(subsample):
    subset = tval[tjf[lab]==1]
    subsets.append(subset)
    data = subset.iloc[:, 1:].mean().sort_values()
    labels = list(data.index)
    widths = data.values
    ax[i].barh(labels, widths, height=0.5)
    
    
plt.show()

```

3.  Strategize sequence interpretation (clean, difference in modifications).  Start by organizing sequences where a categorical == 1.  Then maybe train an LSTM or build a transformer to get a sense of 'how' a lab produces a sequence of given character.
